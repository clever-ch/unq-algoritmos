```python
import numpy as np
import math as m
import random
import timeit
import matplotlib.pyplot as plt
import scipy as sp
import scipy.special
```

```python
def busquedaSecuencial(xs, n):
    for x in xs:
        if(x == n):
            return n
```

```python
def busquedaBinaria(lista, n):
    
    mitad = m.floor(len(lista)/2)
    
    if(len(lista) <= 2):
        return busquedaSecuencial(lista, n)
    else:
        if(n < lista[mitad]):
            return busquedaBinaria(lista[0:mitad], n)
        else:
            return busquedaBinaria(lista[mitad:], n)


```

```python
def busquedaTernaria(lista, n):
    
    unTercioLista = m.floor(len(lista)/3)
    dosTercioLista = unTercioLista * 2
    
    if(len(lista) <= 3):
        return busquedaSecuencial(lista, n)
    else:
        if(n < lista[unTercioLista]):
            return busquedaTernaria(lista[0:unTercioLista], n)
        elif (n < lista[dosTercioLista]):
            return busquedaTernaria(lista[unTercioLista:dosTercioLista], n)
        else:
            return busquedaTernaria(lista[dosTercioLista:], n)


```

```python
n_min = 10
n_max = 10000
st = 10
iteraciones = 10 # iteraciones del benchmark


t_busqueda_binaria = list()
for n in range(n_min, n_max, st):
    
    suma = 0    
    for foo in range(1,iteraciones):
        lista = list(range(1,n))
        suma += timeit.timeit(lambda: busquedaBinaria(lista,n), number = 1)
    promedio = suma / iteraciones
    t_busqueda_binaria.append(promedio)


t_busqueda_ternaria = list()
for n in range(n_min, n_max, st):
    
    suma = 0    
    for foo in range(1,iteraciones):
        lista = list(range(1,n))
        suma += timeit.timeit(lambda: busquedaTernaria(lista,n), number = 1)
    promedio = suma / iteraciones
    t_busqueda_ternaria.append(promedio)



eje_x = list(range(n_min, n_max, st))
plt.plot(eje_x, t_busqueda_binaria, label = 'Búsqueda binaria')
plt.plot(eje_x, t_busqueda_ternaria, label = 'Búsqueda ternaria')

plt.xlabel('n')
plt.ylabel('costo')

plt.legend()
```
